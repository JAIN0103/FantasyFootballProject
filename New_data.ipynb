```python
import bs4 as bs
import urllib.request
import pandas as pd
import numpy as np
URL_list = []
source = urllib.request.urlopen('https://www.fantasypros.com/nfl/reports/leaders/ppr.php?year=2020')
soup = bs.BeautifulSoup(source,'lxml')
for x in soup.find_all("tr"):
    for y in x.find_all("a"):
        url = y.get("href")
        if url[5:12] == "players":
            z = "https://www.fantasypros.com/nfl/stats/" + url[13:]
            URL_list.append(z)
true_URL_list = URL_list[:200]
print(true_URL_list)
```

```python
df_qb = []
df_rb = []
df_wr_te = []
df_QB = pd.DataFrame(df_qb)
df_RB = pd.DataFrame(df_rb)
df_WR_TE = pd.DataFrame(df_wr_te)
for elem in true_URL_list:
    source = urllib.request.urlopen(elem)
    soup = bs.BeautifulSoup(source,'lxml')
    df_list = []
    player_name = elem[38:-4]
    player_name = player_name.replace('-',' ')
    if "Passing Stats" == soup.find_all("th")[0].text:
        for x in soup.find_all("tbody"):
          for y in x.find_all("tr"):
            list = []
            for z in y.find_all("td"):
              list.append(z.text)
            df_list.append(list)
        df = pd.DataFrame(df_list)
        df_passing = df.iloc[:int(len(df.index)/2):]
        df_rushing = df.iloc[int(len(df.index)/2)::]
        new_df = pd.merge(df_passing, df_rushing, on=0)
        for val in range(len(new_df[0])):
            new_df[0][val] = new_df[0][val] + ' ' + player_name
        df_QB = pd.concat([df_QB, new_df])
    if "Rushing Stats" == soup.find_all("th")[0].text:
        for x in soup.find_all("tbody"):
          for y in x.find_all("tr"):
            list = []
            for z in y.find_all("td"):
              list.append(z.text)
            df_list.append(list)
        df = pd.DataFrame(df_list)
        df_rushing = df.iloc[:int(len(df.index)/2):]
        df_receiving = df.iloc[int(len(df.index)/2)::]
        new_df = pd.merge(df_rushing, df_receiving, on=0)
        for val in range(len(new_df[0])):
            new_df[0][val] = new_df[0][val] + ' ' + player_name
        df_RB = pd.concat([df_RB, new_df])
    if "Receiving Stats" == soup.find_all("th")[0].text:
        for x in soup.find_all("tbody"):
          for y in x.find_all("tr"):
            list = []
            for z in y.find_all("td"):
              list.append(z.text)
            df_list.append(list)
        df = pd.DataFrame(df_list)
        df_receiving = df.iloc[:int(len(df.index)/2):]
        df_rushing = df.iloc[int(len(df.index)/2)::]
        new_df = pd.merge(df_receiving, df_rushing, on=0)
        for val in range(len(new_df[0])):
            new_df[0][val] = new_df[0][val] + ' ' + player_name
        df_WR_TE = pd.concat([df_WR_TE, new_df])
print(df_QB)
print(df_RB)
print(df_WR_TE)
```

```python
df_QB.to_excel("Updated_QB_Data.xlsx")
df_RB.to_excel("Updated_RB_Data.xlsx")
df_WR_TE.to_excel("Updated_WR_TE_Data.xlsx")
```
